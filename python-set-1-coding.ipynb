```python
def factorial(n):
    result=1
    for i in range(1,n+1):
        result=result*i
    return result
factorial(5)
        
```

```python
def primenumber(n):
    if n<=1:
        return False
    for i in range(2,int(n*0.5)+1):
        if n%i==0:
            return False
        else:
            return True

if primenumber(7):
    print("Prime number")
else:
    print("Not a primenumber")
    
```
