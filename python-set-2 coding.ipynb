```python
def fibonacci(n):
    sequence=[0,1]
    for i in range(2,n):
        sequence.append(sequence[-1]+sequence[-2])
    return sequence

fibonacci(10)
```

```python
def palindrome(s):
    return s==s[::-1]
word="madam"
if palindrome(word):
    print(word, "is palindrome")
else:
    print(word, "it is not palindrome")
```

```python
def vowelscount(s):
    vowels="aeiouAEIOU"
    return sum(1 for char in s if char in vowels)

print("no.of vowels:",vowelscount("education"))
```

```python
def removeduplicate(input_list):
    return list(set(input_list))

print("removeduplicates:", removeduplicate([1, 2, 2, 3, 4, 4, 5]))
```
